```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"/home/moglix/Downloads/EV_new (2).csv")
```

```python
df
```

## Bar Charts

```python
electric_vehicle_counts = df['Electric Vehicle Type'].value_counts()

plt.figure(figsize=(10, 6)) 
electric_vehicle_counts.plot(kind='bar', color='skyblue')
plt.title("Electric Vehicle Type Distribution")
plt.xlabel("Electric Vehicle Type")
plt.ylabel("Count")
plt.xticks(rotation=45) 
plt.tight_layout()

plt.show()
```

```python
eligibility_counts = df['Clean Alternative Fuel Vehicle (CAFV) Eligibility'].value_counts()
plt.figure(figsize=(10, 6)) 
eligibility_counts.plot(kind='bar', color='green')
plt.title("CAFV Eligibility Distribution")
plt.xlabel("Eligibility Status")
plt.ylabel("Count")
plt.xticks(rotation=45) 
plt.tight_layout()
plt.show()
```

```python
make_counts = df['Make'].value_counts()
plt.figure(figsize=(12, 6)) 
make_counts.plot(kind='bar', color='purple')
plt.title("Car Make Distribution")
plt.xlabel("Make")
plt.ylabel("Count")
plt.xticks(rotation=90) 
plt.tight_layout()
plt.show()
```

## Histogram

```python
electric_range = df['Electric Range']
plt.figure(figsize=(10, 6))  
plt.hist(electric_range, bins=20, color='blue', edgecolor='black')
plt.title("Electric Range Histogram")
plt.xlabel("Electric Range")
plt.ylabel("Frequency")
plt.grid(axis='y', alpha=0.75)
plt.tight_layout()
plt.show()
```

```python
model_year = df['Model Year']
plt.figure(figsize=(10, 6))  
plt.hist(model_year, bins=20, color='orange', edgecolor='black')
plt.title("Model Year Histogram")
plt.xlabel("Model Year")
plt.ylabel("Frequency")
plt.grid(axis='y', alpha=0.75)
plt.tight_layout()
plt.show()
```

## Scatter_plot

```python
electric_range = df['Electric Range']
base_msrp = df['Base MSRP']

plt.figure(figsize=(10, 6)) 
plt.scatter(electric_range, base_msrp, color='red', alpha=0.5)
plt.title("Electric Range vs. Base MSRP")
plt.xlabel("Electric Range")
plt.ylabel("Base MSRP")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
model_year = df['Model Year']
base_msrp = df['Base MSRP']

plt.figure(figsize=(10, 6)) 
plt.scatter(model_year, base_msrp, color='green', alpha=0.5)
plt.title("Model Year vs. Base MSRP")
plt.xlabel("Model Year")
plt.ylabel("Base MSRP")
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Box Plot

```python
plt.figure(figsize=(12, 8)) 
df.boxplot(column='Base MSRP', by='Electric Vehicle Type', vert=False)
plt.title("Base MSRP by Electric Vehicle Type")
plt.xlabel("Base MSRP")
plt.ylabel("Electric Vehicle Type")
plt.grid(axis='x', alpha=0.75)
plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(14, 14))  
plt.xticks(rotation=90)  
plt.boxplot([df[df['Make'] == make]['Electric Range'] for make in df['Make'].unique()], labels = df['Make'].unique())
plt.title("Electric Range by Manufacturer (Make)")
plt.xlabel("Manufacturer (Make)")
plt.ylabel("Electric Range")
plt.grid(axis='y', alpha=0.75)
plt.tight_layout()
plt.show()
```
